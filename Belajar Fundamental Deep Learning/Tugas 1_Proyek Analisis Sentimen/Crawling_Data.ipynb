```python
from google_play_scraper import reviews_all, Sort
import pandas as pd

# Mengambil semua ulasan untuk aplikasi Gojek
result = reviews_all(
    'com.netflix.mediaclient',
    sleep_milliseconds=0, # Atur ke 0 untuk kecepatan maksimal
    lang='id',            # Bahasa ulasan
    country='id',         # Negara target
    sort=Sort.NEWEST    # Urutkan berdasarkan yang terbaru
)

# Mengubah hasil menjadi DataFrame pandas
df_reviews = pd.DataFrame(result)

# Menampilkan 5 baris pertama
print(df_reviews.head())

# Menyimpan ke file CSV
df_reviews.to_csv('ulasan_netflix.csv', index=False)
```
